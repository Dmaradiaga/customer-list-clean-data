```python
import pandas as pd
import re
import numpy as np
```

```python
customers_df = pd.read_excel('Customer Call List.xlsx')
customers_df.tail()
```

```python
#return only valid values
def replace_values(text):
    if text=='N/a':
        return np.nan
    return text
```

```python
df_columns = customers_df.columns
customers_df[df_columns] = customers_df[df_columns].applymap(replace_values)
customers_df
```

```python
#drops duplicates
customers_df = customers_df.drop_duplicates()
customers_df = customers_df.dropna()
customers_df
```

```python
#drop columns 'Do_Not_Contact','Not_Useful_Column'
columns = ['Do_Not_Contact','Not_Useful_Column']
customers_df = customers_df.drop(columns=columns, axis=1)
customers_df
```

```python
#check data types
customers_df.dtypes
```

```python
#drop symbols
def dropSymbols(text):
    if isinstance(text,str):
        return re.sub(r'\W+','',text)
    return text
```

```python
def clearCustomers(columns,df):
    df[columns] = df[columns].applymap(dropSymbols)
```

```python
df_columns = ["First_Name","Last_Name","Phone_Number"]
clearCustomers(df_columns,customers_df)
customers_df
```

```python
#phone number correct format
def format_phone(text):
    return f"{text[:3]}-{text[3:6]}-{text[6:]}"
```

```python
customers_df['Phone_Number'] = customers_df['Phone_Number'].apply(format_phone)
customers_df
```

```python
#changes value fomr Paying Customer
def format_PayingCustomer(text):
    if text=='Yes':
        return 'Y'
    return 'N'
```

```python
customers_df['Paying Customer'] = customers_df['Paying Customer'].apply(format_PayingCustomer)
customers_df.reset_index().drop(columns=['index'], axis=1, inplace=True)
customers_df
```

```python
#save clean data
customers_df.to_excel('Customer-Call-List-clean.xlsx', index=False)
```
